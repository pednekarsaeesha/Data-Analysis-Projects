## Find your next Movie

 __Filter__ the Dataset for movies that meet the following conditions:

__Search 1: Science Fiction Action Movie with Bruce Willis (sorted from high to low Rating)__

```python
mask_genres = df['genres'].str.contains("Action") & df['genres'].str.contains("Science Fiction")
mask_genres
```

```python
actor = df.cast.str.contains('Bruce Willis')
actor
```

```python
df.loc[mask_genres & actor,["title","vote_average"]].sort_values(by="vote_average",ascending = False)
```

__Search 2: Movies with Uma Thurman and directed by Quentin Tarantino (sorted from short to long runtime)__

```python
actor_1 = df['cast'].str.contains('Uma Thurman')
actor_1
```

```python
director_1 = df['director'].str.contains('Quentin Tarantino')
director_1
```

```python
df.loc[actor_1 & director_1,['title','runtime']].sort_values(by='runtime')
```

__Search 3: Most Successful Pixar Studio Movies between 2010 and 2015 (sorted from high to low Revenue)__

```python
df.production_companies.str.contains('Pixar')
```

```python
pixar = df.production_companies.str.contains('Pixar').fillna(False)
pixar
```

```python
time = df['release_date'].between("2010-01-01","2015-12-31")
time
```

```python
df.loc[pixar & time,['title','production_companies','release_date','revenue_musd']].sort_values(by="revenue_musd",ascending=False)
```

__Search 4: Action or Thriller Movie with original language English and minimum Rating of 7.5 (most recent movies first)__

```python
movie_genre = df['genres'].str.contains('Action') | df['genres'].str.contains('Thriller')
movie_genre
```

```python
language = df['original_language'].str.contains('en')
language
```

```python
ratings = df['vote_average']>=7.5
```

```python
df.loc[movie_genre & language &ratings, ['title','genres','original_language','release_date','vote_average','vote_count']].sort_values(by="release_date",ascending=False)
```

## Are Franchises more successful?

 __Analyze__ the Dataset and __find out whether Franchises (Movies that belong to a collection) are more successful than stand-alone movies__ in terms of:

- mean revenue
- median Return on Investment
- mean budget raised
- mean popularity
- mean rating

__Franchise vs. Stand-alone: Average Revenue__

```python
df['belongs_to_collection'].value_counts(dropna=False)
```

```python
df['Franchise']=df['belongs_to_collection'].notna()
```

```python
df['Franchise'].value_counts()
```

```python
df.groupby("Franchise").mean()
```

__Franchise vs. Stand-alone: Return on Investment / Profitability (median)__

```python
df['ROI'] = df['revenue_musd'].div(df['budget_musd'])
```

```python
df.groupby('Franchise')['ROI'].median()
```

__Franchise vs. Stand-alone: Average Popularity__

```python
df.groupby('Franchise')['popularity'].mean()
```

__Franchise vs. Stand-alone: Average Budget__

```python
df.groupby('Franchise')['budget_musd'].mean()
```

__Franchise vs. Stand-alone: Average Rating__

```python
df.groupby("Franchise")['vote_average'].mean()
```

```python
df.groupby('Franchise').agg({"budget_musd":"mean", "revenue_musd":"mean",
                             "vote_average":"mean","popularity":"mean","ROI":"median","vote_count":"mean"})
```

## Most Successful Franchises

 __Find__ the __most successful Franchises__ in terms of

- __total number of movies__
- __total & mean budget__
- __total & mean revenue__
- __mean rating__

```python
df.belongs_to_collection.value_counts(dropna=False)
```

```python
franchises = df.groupby("belongs_to_collection").agg({"title":"count", "budget_musd": ["sum", "mean"], 
                                                      "revenue_musd": ["sum", "mean"],
                                                      "vote_average": "mean", "popularity": "mean",
                                                      "ROI":"median", 
                                                      "vote_count":"mean"})
franchises
```

```python
franchises.nlargest(10,("title","count"))
```

```python
franchises.nlargest(20, ("revenue_musd",'mean'))
```

```python
franchises.nlargest(20,("revenue_musd","sum"))
```

```python
franchises.nlargest(20,("budget_musd","mean"))
```

```python
franchises.nlargest(20,("budget_musd","sum"))
```

```python
franchises[franchises[("vote_count","mean")]>=1000].nlargest(20,("vote_average","mean"))
```

## Most Successful Directors

 __Find__ the __most successful Directors__ in terms of

- __total number of movies__
- __total revenue__
- __mean rating__

```python
df.director
```

```python
df.director.value_counts().head(20)
```

```python
directors = df.groupby('director').agg({"title":"count","revenue_musd":"sum",
                            "vote_average":"mean","vote_count":"sum"})
directors
```

```python
directors.nlargest(20,"revenue_musd")
```

```python
directors[(directors.vote_count>10000) & (directors.title>=10)].nlargest(20,"vote_average")
```
